```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
import numpy as np
button_routes = np.arange(0,1001*40,step=40)
button_routes
```

```python
import pandas as pd
```

```python
result = []
```

```python
driver:webdriver.Chrome = webdriver.Chrome('/home/admin/Desktop/bagsnlp/Javi-MachineTranslation/chromedriver')
options = webdriver.ChromeOptions() 
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
chrome_prefs = {}
options.experimental_options["prefs"] = chrome_prefs
chrome_prefs["profile.default_content_settings"] = {"images": 2}
for button_route in button_routes:
    driver.get('https://www.opensubtitles.org/en/search/sublanguageid-jpn,vie/offset-'+str(button_route))    
    elems = driver.find_elements(By.CLASS_NAME, 'change')
    for elem in elems:
        tds = elem.find_elements(By.TAG_NAME,'td')
        film_name = tds[0].find_elements(By.TAG_NAME,'strong')[0].text
        flag = tds[1].find_elements(By.CLASS_NAME,'flag')[0].get_attribute('class')
        # film_name = elem.find_elements(By.TAG_NAME,'strong')[0].text
        # flag = elem.find_elements(By.CLASS_NAME,'flag')[0].get_attribute('class')
        link = tds[4].find_elements(By.TAG_NAME,'a')[0].get_attribute('href')
        result.append([film_name,flag.replace('flag ',''),link])
driver.close()
```

```python
result
```

```python
df = pd.DataFrame(result, columns=['film_name','flag', 'link'])
```

```python
df.to_csv('data.csv',index=False)
```
